```python
import pandas as pd

# Assuming you have loaded 'movies' and 'ratings' DataFrames correctly
movies = pd.read_csv('tmdb_5000_movies.csv')
ratings = pd.read_csv('ratings.csv')

# Print column names to identify the exact names
print("Movies columns:", movies.columns)
print("Ratings columns:", ratings.columns)

# Adjust column names for merging
movies.rename(columns={'id': 'movie_id'}, inplace=True)
ratings.rename(columns={'movieid': 'movie_id'}, inplace=True)

# Merge based on common column 'movie_id'
merged_df=movies
merged_df=merged_df.merge(ratings,on='movie_id')
```

```python
import pandas as pd
from surprise import Dataset, Reader, KNNBasic
from surprise.model_selection import train_test_split
from surprise.accuracy import rmse
```

```python
reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(merged_df[['userId', 'movie_id', 'rating']], reader)
```

```python
merged_df.info()
```

```python
trainset, testset = train_test_split(data, test_size=0.2, random_state=42)
```

```python
print (data)
```

```python
model = KNNBasic(sim_options={'user_based': True})
model.fit(trainset)
```

```python
predictions = model.test(testset)
accuracy = rmse(predictions)
print(f'RMSE on test set: {accuracy}')
```

```python
unique_movies = merged_df['title'].unique()

for movie_title in unique_movies:
    print(f"Recommendations for {movie_title}:")
    movie_ids = merged_df[merged_df['title'] == movie_title]['movie_id'].tolist()
    print(movie_ids)
    for movie_id in movie_ids:
        try:
            # Check if movie_id exists in the trainset
            if movie_id in model.trainset._raw2inner_id_items:
                movie_inner_id = model.trainset.to_inner_iid(movie_id)
                if movie_inner_id is not None:
                    similar_movies = model.get_neighbors(movie_inner_id, k=5)
                    for similar_inner_id in similar_movies:
                        similar_movie_id = model.trainset.to_raw_iid(similar_inner_id)
                        recommended_movie_title = merged_df[merged_df['movie_id'] == similar_movie_id]['title'].iloc[0]
                        print(recommended_movie_title)
                else:
                    print(f"Warning: Movie with movie_id {movie_id} has no inner id.")
            else:
                print(f"Warning: Movie with movie_id {movie_id} not found in the training set.")
        except IndexError:
            print(f"Error: Index error for movie_id {movie_id}.")
        
    print("-------------------")
```

```python
import pandas as pd
from sklearn.neighbors import NearestNeighbors

# Assuming merged_df contains userId, movie_id, title, and rating columns
# Print unique movie titles to verify
print("Unique movie titles in the dataset:")
print(merged_df['title'].unique())

# Create a pivot table (movie_user_matrix)
movie_user_matrix = merged_df.pivot_table(index='movie_id', columns='userId', values='rating').fillna(0)

# Transpose the movie_user_matrix for item-based collaborative filtering
movie_movie_matrix = movie_user_matrix.T

# Fit the KNN model using cosine similarity
model_knn_item = NearestNeighbors(metric='cosine', algorithm='brute')
model_knn_item.fit(movie_movie_matrix.values)

# Define the recommendation function
def recommend_item_based(movie_title):
    # Find the movie_id for the given movie title
    movie_ids = merged_df[merged_df['title'] == movie_title]['movie_id'].unique()
    if len(movie_ids) == 0:
        print(f"Movie '{movie_title}' not found in the dataset.")
        return
    
    movie_id = movie_ids[0]  # Take the first matching movie_id
    if movie_id not in movie_user_matrix.index:
        print(f"Movie '{movie_title}' not found in the training set.")
        return
    
    movie_index = movie_user_matrix.index.get_loc(movie_id)
    distances, indices = model_knn_item.kneighbors(movie_movie_matrix.iloc[movie_index, :].values.reshape(1, -1), n_neighbors=6)
    
    print(f"Recommendations for '{movie_title}':")
    for i in range(1, len(distances.flatten())):
        recommended_movie_index = indices.flatten()[i]
        recommended_movie_id = movie_user_matrix.index[recommended_movie_index]
        recommended_movie_title = merged_df[merged_df['movie_id'] == recommended_movie_id]['title'].iloc[0]
        print(recommended_movie_title)

# Example usage
recommend_item_based('Minority Report')
```

```python
def recommend(movie):
    try:
        # Clean up input movie title for consistency
        movie = movie.strip().lower()  # Remove leading/trailing spaces and convert to lowercase
        
        # Check if movie exists in the movies DataFrame
        if movie not in merged_df['title'].str.strip().str.lower().values:
            print(f"Movie '{movie}' not found in the dataset.")
            return
        
        # Find movie index in merged_df
        movie_index = merged_df.index[merged_df['title'].str.strip().str.lower() == movie]
        
        if len(movie_index) == 0:
            raise IndexError(f"Index for '{movie}' not found in merged_df.")
        
        movie_index = movie_index[0]
        
        # Debug print to check movie_index
        print(f"Movie index for '{movie}': {movie_index}")
        
        # Check if movie_index is within valid range for similarity matrix
        if movie_index >= len(similarity) or movie_index < 0:
            raise IndexError(f"Index {movie_index} is out of bounds for similarity matrix.")
        
        distances = similarity[movie_index]
        
        # Debug print to check distances
        print(f"Distances for '{movie}': {distances}")
        
        # Sort distances and retrieve top recommendations
        movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
        
        print(f"Recommendations for '{movie}':")
        for i, (idx, sim) in enumerate(movies_list, 1):
            if idx < len(merged_df):
                recommended_movie_title = merged_df.iloc[idx]['title']
                print(f"{i}. {recommended_movie_title}")
            else:
                print(f"Index {idx} is out of bounds for DataFrame.")
    
    except IndexError as e:
        print(f"Error in recommend('{movie}'): {str(e)}")
    except Exception as e:
        print(f"Unexpected error in recommend('{movie}'): {str(e)}")

# Example usage
recommend('silent running')
```
